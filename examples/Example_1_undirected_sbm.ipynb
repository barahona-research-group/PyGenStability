# Example 1: Markov Stability applied to undirected SBM

```python
import networkx as nx

from pygenstability import run, plotting

import create_graph
```

```python
# creating SBM

nx_graph, adjacency = create_graph.create_sbm()
```

```python
# run Markov Stability
all_results = run(adjacency, min_scale=-1, max_scale = 1, n_scale = 30, constructor='continuous_normalized')
```

```python
# plot the results matplotlib
_ = plotting.plot_scan(all_results, use_plotly=False)
```

```python
# plot partition at scale index t
plotting.plot_single_partition(nx_graph, all_results, scale_id=20)
```

```python
# plot optimal partitions
plotting.plot_optimal_partitions(nx_graph,all_results)
```
