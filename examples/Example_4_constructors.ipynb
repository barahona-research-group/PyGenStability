# Example 4: Trying various constructor options

Each constructor encodes a different form of dynamics on the graph. Here, we exemplify each constructor and their outputs.

```python
import networkx as nx

from pygenstability import run, plotting
import create_graph
```

```python
# creating SBM

nx_graph, adjacency = create_graph.create_sbm()
```

## Linearised matrix exponential (modularity)

```python
all_results = run(adjacency, min_scale=-1, max_scale=1, n_scale=30, constructor='linearized')
_ = plotting.plot_scan(all_results)
```

## Continuous scale - combinatorial Laplacian

```python
all_results = run(adjacency, min_scale=-0.5, max_scale =.2, n_scale=30, constructor='continuous_combinatorial')
_ = plotting.plot_scan(all_results)
```

## Continuous scale - normalized Laplacian

```python
all_results = run(adjacency, min_scale=-1, max_scale=1, n_scale=30, constructor='continuous_normalized')
_ = plotting.plot_scan(all_results)
```
